```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../03.신영화/data/납 월별 도시별 대기오염도.xlsx', header=None)
df.head(5)
```

```python
df = df.rename(columns=df.iloc[0])        #첫번째 행을 column으로 지정
df.head(3)
```

```python
df = df.drop(df.index[0])        #첫번째 행을 삭제한다
df
```

```python
df.fillna(0, inplace=True)
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
df
```

```python
df.mean()
df
```

```python
df.loc['평균'] = df.mean()
df
```
