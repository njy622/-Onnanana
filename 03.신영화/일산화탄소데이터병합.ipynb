```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CSV 파일에서 데이터를 불러오기
df = pd.read_csv('data/질병성분데이터.csv', encoding='utf-8')
df.head(3)
```

```python
transposed_df = df.transpose()
transposed_df.head(3)
```

```python
df = transposed_df.rename(columns=transposed_df.iloc[0])
df.head(3)
```

```python
df = df.drop(df.index[0])

df.head(3)
```

```python
# 선택한 컬럼만 추출
selected_columns = ['일산화탄소', '심근경색', '폐질환', '뇌졸증', '천식','폐렴']
df_selected = df[selected_columns]
df_selected.tail(3)
```

```python
# 일산화탄소가 높을 때의 데이터와 낮을 때의 데이터로 나누기
high_pm_data = df_selected[df_selected['일산화탄소'] > df_selected['일산화탄소'].median()]
low_pm_data = df_selected[df_selected['일산화탄소'] <= df_selected['일산화탄소'].median()]
```

```python
# 높은 일산화탄소 그룹과 낮은 일산화탄소 그룹에서의 평균 비교
impact_percentages = ((high_pm_data.mean() - low_pm_data.mean()) / low_pm_data.mean()) * 100
```

```python
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
# 모든 값을 양수로 만들기
impact_percentages = impact_percentages.abs()
```

```python
import seaborn as sns

colors = sns.color_palette("Pastel1")

# 파이차트 그리기
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(impact_percentages[1:], labels=impact_percentages.index[1:], autopct='%1.0f%%', startangle=90, textprops={'fontsize': 10}, radius=1.2, pctdistance=0.85, labeldistance=1.03, colors=colors, explode=[0,0,0,0,0.07],shadow=True)

# 텍스트 자동 조정 및 위치 조정
for autotext in autotexts:
    autotext.set(size=8, ha='center', va='center', rotation_mode='anchor')

ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('일산화탄소 영향')
plt.show()
```

```python
# 막대 그래프 그리기
fig, ax = plt.subplots()
impact_percentages[1:].plot(kind='bar', ax=ax, color='skyblue')
ax.set_ylabel('영향을 받는 비율 (%)')
ax.set_xlabel('컬럼')
ax.set_title('일산화탄소')

plt.show()
```
