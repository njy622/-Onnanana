```python
import pandas as pd
```

```python
df = pd.read_excel('data/환경성질환_천식.xlsx', header=None)
df.head(5)
```

```python
# 진료년월 행을 컬럼으로 지정
new_header = df.iloc[2]
df = df[3:]  # 3번째 행부터 데이터로 사용
df.columns = new_header
df.head(5)
```

```python
# NaN 값을 다른 값으로 대체합니다.
df.columns = df.columns.fillna('')

# "월"이라는 단어가 포함된 컬럼만 남기고 나머지는 삭제합니다.
df = df[df.columns[df.columns.str.contains('월')]]
```

```python
df
```

```python
# 첫 번째 열을 인덱스로 설정합니다.
df = df.set_index(df.columns[0])
```

```python
df.head(5)
```

```python
# "진료년월"이라는 이름의 인덱스를 삭제합니다.
df = df.reset_index(drop=True)
```

```python
df
```

```python
# 두 번째 행만 남기고 나머지 행을 삭제합니다.
df = df.iloc[[1]]
df
```

```python
# 변경된 데이터프레임을 'output.csv' 파일로 저장합니다.
df.to_csv('환경성질환_천식.csv', index=False)
```
