```python
import pandas as pd
```

```python
df = pd.read_excel('data/이산화질소_월별_도시별_대기오염도_20231215115337.xlsx', header=None)
df.head(5)
```

```python
#  구분(2) 행을 컬럼으로 지정
new_header = df.iloc[0]
df = df[1:]  # 3번째 행부터 데이터로 사용
df.columns = new_header
df.head(5)
```

```python
# 결측치 확인
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# 결측치를 0으로 대체
df_filled = df.fillna(0, inplace=True)

# 결과 출력
print("결측치를 0으로 대체한 데이터프레임:")
print(df_filled)
```

```python
# 년도와 월을 나타내는 컬럼명 리스트 생성
columns_to_keep = [f"구분(2)"] + [f"{year}.{month:02d}" for year in range(2018, 2023) for month in range(1, 13)]

# 원하는 컬럼만 남기기
df_filtered = df[columns_to_keep]

# '구분(2)' 컬럼을 이미 인덱스로 가지고 있는지 확인
if '구분(2)' in df_filtered.columns:
    # 이미 인덱스로 설정되어 있다면, 인덱스를 해제하고 컬럼 필터링
    df_filtered = df_filtered.reset_index().loc[:, columns_to_keep].set_index('구분(2)')


df_filtered.head(5)
```

```python
# 년도와 월을 나타내는 컬럼명 리스트 생성
columns_to_keep = [f"구분(2)"] + [f"{year}.{month:02d}" for year in range(2018, 2023) for month in range(1, 13)]

# 원하는 컬럼만 남기기
df_filtered = df[columns_to_keep]

# '구분(2)' 컬럼을 인덱스로 설정
df_filtered.set_index('구분(2)', inplace=True)

df_filtered.head(5)

```
