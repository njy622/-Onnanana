```python
import requests, json
import pandas as pd
from urllib.parse import quote
```

```python
df = pd.read_csv("data/측정소좌표.csv", index_col="Unnamed: 0")
df
```

```python
gu = pd.read_csv("data/위도경도_최종.csv")
gu
```

```python
juso = df.주소.to_list()
juso
```

```python
gu = gu.시군구.to_list()
gu
```

```python
juso[100].split()
```

```python
# '구' 열을 초기화합니다.
df['구'] = None

for i in range(len(juso)):
    address_split = juso[i].split()  # i번째 주소를 리스트로 분리
    for word in address_split:
        for g in gu:            # 시군구 리스트를 하나씩 반복해서
            if word in g:       # 분리한 리스트 주소에 해당 시군구가 있는지 체크
                df.at[i, '구'] = g  # 해당 행의 '구' 열에 행정동명을 추가합니다.
                break   # 해당 주소의 '구' 열이 채워졌으면 더 이상 시군구를 체크할 필요가 없으므로 반복문을 종료합니다.
df
```

```python
df3 = df.drop_duplicates(subset="구")
df3
```

```python
df3.reset_index(inplace=True)
```

```python
import folium

map = folium.Map(location=[37.56682, 126.97865], zoom_start=13)  # 중심 위치

for i in df3.index:
    if i in df3.index and pd.notnull(df3['위도'][i]) and pd.notnull(df3['경도'][i]):  # 해당 인덱스가 있는지 체크 후 None 값인 경우 건너뜀
        folium.Marker(
            location=[df3['위도'][i], df3['경도'][i]],  # WGS84 좌표계: 위도(latitude), 경도(longitude)
            tooltip=df3['측정소명'][i],
            popup=folium.Popup(df3['주소'][i], max_width=200)
        ).add_to(map)

title_html = '<h3 align="center" style="font-size:20px">측정소 위치</h3>'
map.get_root().html.add_child(folium.Element(title_html))

map
```

```python
df3.to_csv("data/측정소좌표(시군구구분_중복제거).csv")
df3
```
