```python
import matplotlib.pyplot as plt
import seaborn as sns
# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 년월을 datetime 객체로 변환
df['STD_YYYYMM'] = pd.to_datetime(df['STD_YYYYMM'], format='%Y%m')

# 전국 평균 미세먼지 농도 계산
df_avg = df.groupby('STD_YYYYMM')['차량수'].mean().reset_index()

# 코로나19 발생 시기를 표시하기 위한 범위 설정 (예: 2020년 1월부터 2021년 12월까지)
start_date = '2020-01-01'
end_date = '2021-12-31'

# 필터링된 데이터 프레임 생성
df_filtered = df_avg[(df_avg['STD_YYYYMM'] >= start_date) & (df_avg['STD_YYYYMM'] <= end_date)]

# 시각화
plt.figure(figsize=(12, 6))
sns.lineplot(x='STD_YYYYMM', y='차량수', data=df_filtered, marker='o')

plt.title('코로나19 기간 동안 년월별 전국 차량수')
plt.xlabel('년월')
plt.ylabel('차량수')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 년월을 datetime 객체로 변환
dfsm['STD_YYYYMM'] = pd.to_datetime(dfsm['STD_YYYYMM'], format='%Y%m')

# 전국 평균 미세먼지 및 환자 수 계산
dfsm_avg = dfsm.groupby('STD_YYYYMM').agg({
    '차량수': 'mean',
    '아토피환자수': 'sum',   # 아토피환자 수
    'AVG_구리': 'sum'  # 구리
}).reset_index()

# Min-Max 스케일링
scaler = MinMaxScaler()
dfsm_avg[['차량수', '아토피환자수', 'AVG_구리']] = scaler.fit_transform(dfsm_avg[['차량수', '아토피환자수', 'AVG_구리']])

# 코로나19 이전과 이후 기간 설정
pre_covid_start = '2010-01-01'
pre_covid_end = '2019-12-31'
covid_start = '2020-01-01'
covid_end = '2021-12-31'

# 필터링된 데이터 프레임 생성
dfsm_pre_covid = dfsm_avg[(dfsm_avg['STD_YYYYMM'] >= pre_covid_start) & (dfsm_avg['STD_YYYYMM'] <= pre_covid_end)]
dfsm_covid = dfsm_avg[(dfsm_avg['STD_YYYYMM'] >= covid_start) & (dfsm_avg['STD_YYYYMM'] <= covid_end)]

# 변수별 평균값 계산
car_pre_covid = dfsm_pre_covid['차량수'].mean()
car_covid = dfsm_covid['차량수'].mean()
atopy_pre_covid = dfsm_pre_covid['아토피환자수'].mean()
atopy_covid = dfsm_covid['아토피환자수'].mean()
cu_pre_covid = dfsm_pre_covid['AVG_구리'].mean()
cu_covid = dfsm_covid['AVG_구리'].mean()


# 첫 번째 그래프: 미세먼지와 아토피환자수
fig, ax1 = plt.subplots(figsize=(15, 8))
sns.lineplot(x='STD_YYYYMM', y='차량수', data=dfsm_avg, marker='o', color='blue', label='전국차량수', ax=ax1)
sns.lineplot(x='STD_YYYYMM', y='아토피환자수', data=dfsm_avg, marker='o', color='green', label='아토피환자 수', ax=ax1)

# 코로나19 이전 평균선
ax1.axhline(car_pre_covid, color='blue', linestyle='--', xmin=0, xmax=0.48, label='전국차량수 평균')
ax1.axhline(atopy_pre_covid, color='green', linestyle='--', xmin=0, xmax=0.48, label='아토피환자수 평균')

# 코로나19 이후 평균선
ax1.axhline(car_covid, color='blue', linestyle='--', xmin=0.52, xmax=1)
ax1.axhline(atopy_covid, color='green', linestyle='--', xmin=0.52, xmax=1)


ax1.set_title('년월별 전국차량수 및 아토피환자 수')
ax1.set_ylabel('스케일링된 값')
ax1.legend()

# 두 번째 그래프: 전국차량수와 AVG_구리
fig, ax2 = plt.subplots(figsize=(15, 8))
sns.lineplot(x='STD_YYYYMM', y='차량수', data=dfsm_avg, marker='o', color='blue', label='전국차량수', ax=ax2)
sns.lineplot(x='STD_YYYYMM', y='AVG_구리', data=dfsm_avg, marker='o', color='purple', label='구리', ax=ax2)

# 코로나19 이전 평균선
ax2.axhline(car_pre_covid, color='blue', linestyle='--', xmin=0, xmax=0.48, label='전국차량수 평균')
ax2.axhline(cu_pre_covid, color='purple', linestyle='--', xmin=0, xmax=0.48, label='구리 평균')

# 코로나19 이후 평균선
ax2.axhline(car_covid, color='blue', linestyle='--', xmin=0.52, xmax=1)
ax2.axhline(cu_covid, color='purple', linestyle='--', xmin=0.52, xmax=1)



ax2.set_title('년월별 전국차량수 및 구리')
ax2.set_ylabel('스케일링된 값')
ax2.legend()

plt.show()
```

## 가설검정(2-sample T test)

```python
import scipy as sp 
sp.stats.ttest_ind(dfsm[dfsm.구리_BAD=='Y'].차량수, dfsm[dfsm.구리_BAD=='N'].차량수, equal_var=False)
```

 주요 두 부분인 통계적 유의성(statistic)과 p-값(p-value)을 살펴보겠습니다:

통계적 유의성 (statistic = 14.024747907711108):

이 값은 두 그룹 간 평균의 차이가 얼마나 큰지를 나타내는 t-통계량입니다.
값 14.02는 두 그룹 간에 상당한 평균 차이가 있음을 나타냅니다. 일반적으로, t-통계량이 클수록 두 집단 간의 차이가 더 명확하다고 해석됩니다.
p-값 (p-value = 1.0195507152609119e-38):

p-값은 귀무 가설(여기서는 두 그룹 간의 평균이 같다는 가설)이 참일 경우, 관찰된 결과(또는 더 극단적인 결과)가 발생할 확률을 나타냅니다.
이 경우의 p-값은 매우 작습니다. 일반적으로 p-값이 0.05 이하일 때, 결과를 통계적으로 유의미하다고 간주합니다. 따라서, 이 p-값은 두 그룹 간에 통계적으로 매우 유의미한 차이가 있음을 나타냅니다.
결론적으로, 이 t-검정 결과는 두 그룹 간에 통계적으로 매우 유의미한 차이가 있음을 나타내며, 이 차이는 상당히 크다고 해석됩니다. 연구의 맥락에서 이 결과는 두 조건이나 그룹 간에 중요한 차이가 있음을 나타내는 강력한 증거로 사용될 수 있습니다.







```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 예제 데이터
group1 = dfsm[dfsm.구리_BAD == 'Y'].차량수
group2 = dfsm[dfsm.구리_BAD == 'N'].차량수

# 평균과 표준 오차 계산
means = [np.mean(group1), np.mean(group2)]
errors = [np.std(group1) / np.sqrt(len(group1)), np.std(group2) / np.sqrt(len(group2))]

# 막대 그래프 생성
labels = ['구리_BAD Y', '구리_BAD N']
plt.bar(labels, means, yerr=errors, capsize=10)
plt.ylabel('차량 수')
plt.title('구리_BAD에 따른 차량 수의 평균 비교')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 데이터프레임 'dfsm'에서 클러스터링에 사용할 변수 선택
X = dfsm[['차량수', 'AVG_구리']]

# K-평균 클러스터링 수행 (클러스터 수는 예제로 2로 설정)
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
labels = kmeans.labels_

# 클러스터링 결과 시각화
plt.scatter(dfsm['차량수'], dfsm['AVG_구리'], c=labels, cmap='viridis', marker='o')
plt.xlabel('차량 수')
plt.ylabel('구리')
plt.title('차량 수와 구리에 따른 클러스터링')
plt.colorbar(label='Cluster Label')
plt.show()
```
