```python
import pandas as pd
```

```python
df = pd.read_csv('../data/일산화탄소측정소.csv', encoding='ansi')
```

```python
df.head(2)
```

```python
long_df = df.melt(id_vars=['구분', '항목','단위'], 
                  var_name='STD_YYYYMM', 
                  value_name='일산화')
```

```python
long_df.head(2)
```

```python
long_df.rename(columns={'구분': 'SIDO'}, inplace=True)
```

```python
# '오창읍' 데이터만 필터링
ochang_df = long_df[long_df['SIDO'] == '송정동']
```

```python
# 출력할 최대 행 수를 설정
pd.set_option('display.max_rows', None)

# 데이터 프레임 출력
print(ochang_df)
```
