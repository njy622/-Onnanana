```python
# 'STD_YYYYMM' 열의 숫자가 아닌 문자를 제거
long_df['STD_YYYYMM'] = long_df['STD_YYYYMM'].str.replace('[^\d]', '', regex=True)
```

```python
# 결과를 정수 타입으로 변환
long_df['STD_YYYYMM'] = long_df['STD_YYYYMM'].astype(int)
```

```python
# SIDO, STD_YYYYMM 열을 기준으로 그룹바이 한후 PM10의 평균값을 새로운 콜롬에 저장
long_df['평균최고기온'] = long_df.groupby(['SIDO', 'STD_YYYYMM'])['평균최고기온'].transform('mean').round(1)
```

```python
df2 = long_df.drop_duplicates(subset=['SIDO', 'STD_YYYYMM', '평균최고기온'])
# SIDO와 STD_YYYYMM 열을 기준으로 중복을 제거
# 중복된 데이터를 제거하면 데이터가 정렬됨 
```

```python
#  항목 , 단위 콜롬 제거
df2 = df2.drop(columns=['항목', '단위'])
```
