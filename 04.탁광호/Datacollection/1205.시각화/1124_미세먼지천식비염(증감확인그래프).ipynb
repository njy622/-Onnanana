```python
import matplotlib.pyplot as plt
import seaborn as sns
# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 년월을 datetime 객체로 변환
df['STD_YYYYMM'] = pd.to_datetime(df['STD_YYYYMM'], format='%Y%m')

# 전국 평균 미세먼지 농도 계산
df_avg = df.groupby('STD_YYYYMM')['AVG_PM10'].mean().reset_index()

# 코로나19 발생 시기를 표시하기 위한 범위 설정 (예: 2020년 1월부터 2021년 12월까지)
start_date = '2020-01-01'
end_date = '2021-12-31'

# 필터링된 데이터 프레임 생성
df_filtered = df_avg[(df_avg['STD_YYYYMM'] >= start_date) & (df_avg['STD_YYYYMM'] <= end_date)]

# 시각화
plt.figure(figsize=(12, 6))
sns.lineplot(x='STD_YYYYMM', y='AVG_PM10', data=df_filtered, marker='o')

plt.title('코로나19 기간 동안 년월별 전국 평균 미세먼지 농도')
plt.xlabel('년월')
plt.ylabel('평균 미세먼지 농도 (μg/m³)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 년월을 datetime 객체로 변환
dfsm['STD_YYYYMM'] = pd.to_datetime(dfsm['STD_YYYYMM'], format='%Y%m')

# 전국 평균 미세먼지 농도 및 환자 수 계산
dfsm_avg = dfsm.groupby('STD_YYYYMM').agg({
    'AVG_PM10': 'mean',
    '천식환자수': 'sum',   # 천식환자 수
    '비염환자수': 'sum'  # 비염환자 수
}).reset_index()

# Min-Max 스케일링
scaler = MinMaxScaler()
dfsm_avg[['AVG_PM10', '천식환자수', '비염환자수']] = scaler.fit_transform(dfsm_avg[['AVG_PM10', '천식환자수', '비염환자수']])

# 코로나19 이전과 이후 기간 설정
pre_covid_start = '2010-01-01'
pre_covid_end = '2019-12-31'
covid_start = '2020-01-01'
covid_end = '2021-12-31'

# 필터링된 데이터 프레임 생성
dfsm_pre_covid = dfsm_avg[(dfsm_avg['STD_YYYYMM'] >= pre_covid_start) & (dfsm_avg['STD_YYYYMM'] <= pre_covid_end)]
dfsm_covid = dfsm_avg[(dfsm_avg['STD_YYYYMM'] >= covid_start) & (dfsm_avg['STD_YYYYMM'] <= covid_end)]

# 변수별 평균값 계산
avg_pm10_pre_covid = dfsm_pre_covid['AVG_PM10'].mean()
avg_pm10_covid = dfsm_covid['AVG_PM10'].mean()
avg_asthma_pre_covid = dfsm_pre_covid['천식환자수'].mean()
avg_asthma_covid = dfsm_covid['천식환자수'].mean()
avg_rhinitis_pre_covid = dfsm_pre_covid['비염환자수'].mean()
avg_rhinitis_covid = dfsm_covid['비염환자수'].mean()


# 첫 번째 그래프: 미세먼지와 천식환자수
fig, ax1 = plt.subplots(figsize=(15, 8))
sns.lineplot(x='STD_YYYYMM', y='AVG_PM10', data=dfsm_pre_covid, marker='o', color='blue', label='미세먼지 (코로나19 이전)', ax=ax1)
sns.lineplot(x='STD_YYYYMM', y='AVG_PM10', data=dfsm_covid, marker='o', color='orange', label='미세먼지 (코로나19 기간)', ax=ax1)
sns.lineplot(x='STD_YYYYMM', y='천식환자수', data=dfsm_pre_covid, marker='o', color='green', label='천식환자 수 (코로나19 이전)', ax=ax1)
sns.lineplot(x='STD_YYYYMM', y='천식환자수', data=dfsm_covid, marker='o', color='red', label='천식환자 수 (코로나19 기간)', ax=ax1)

# 평균선 추가
ax1.axhline(avg_pm10_pre_covid, color='blue', linestyle='--', label='미세먼지 평균 (코로나19 이전)')
ax1.axhline(avg_pm10_covid, color='orange', linestyle='--', label='미세먼지 평균 (코로나19 기간)')
ax1.axhline(avg_asthma_pre_covid, color='green', linestyle='--')
ax1.axhline(avg_asthma_covid, color='red', linestyle='--')


ax1.set_title('년월별 미세먼지 농도 및 천식환자 수')
ax1.set_ylabel('스케일링된 값')
ax1.legend()

# 두 번째 그래프: 미세먼지와 비염환자수
fig, ax2 = plt.subplots(figsize=(15, 8))
sns.lineplot(x='STD_YYYYMM', y='AVG_PM10', data=dfsm_pre_covid, marker='o', color='blue', label='미세먼지 (코로나19 이전)', ax=ax2)
sns.lineplot(x='STD_YYYYMM', y='AVG_PM10', data=dfsm_covid, marker='o', color='orange', label='미세먼지 (코로나19 기간)', ax=ax2)
sns.lineplot(x='STD_YYYYMM', y='비염환자수', data=dfsm_pre_covid, marker='o', color='purple', label='비염환자 수 (코로나19 이전)', ax=ax2)
sns.lineplot(x='STD_YYYYMM', y='비염환자수', data=dfsm_covid, marker='o', color='pink', label='비염환자 수 (코로나19 기간)', ax=ax2)

# 평균선 추가
ax2.axhline(avg_pm10_pre_covid, color='blue', linestyle='--', label='미세먼지 평균 (코로나19 이전)')
ax2.axhline(avg_pm10_covid, color='orange', linestyle='--', label='미세먼지 평균 (코로나19 기간)')
ax2.axhline(avg_rhinitis_pre_covid, color='purple', linestyle='--')
ax2.axhline(avg_rhinitis_covid, color='pink', linestyle='--')


ax2.set_title('년월별 미세먼지 농도 및 비염환자 수')
ax2.set_ylabel('스케일링된 값')
ax2.legend()

plt.show()
```
