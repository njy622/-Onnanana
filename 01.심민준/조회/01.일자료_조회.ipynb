```python
import requests

def weather_data():
    results_list = []
    
    now = datetime.now() # 현재 시각 데이터 생성
    endDt = (now-timedelta(days=1)).strftime('%Y%m%d') # 하루 전 데이터
    
    params = {
        'serviceKey': service_key,
        'numOfRows': 31, # 한달 31일
        'pageNo0': 1,
        'dataType': 'JSON',
        'dataCd': 'ASOS',   # 자료 코드
        'dateCd': 'DAY',    # 날짜 코드
        'startDt': '20231201',  # 시작일 (YYYYMMDD)
        'endDt': endDt,    # 종료일 (HHMM)
        'stnIds': '108'     # 지점 번호 ex)서울
    }

    res = requests.get(url=base_url, params=params)
    data = res.json()['response']['body']['items']['item']

    for item in data:
        results = {
            '일시': item['tm'],
            '지점명': item['stnNm'],
            '평균기온': item['avgTa'],
            '최저기온': item['minTa'],
            '최고기온': item['maxTa'],
            '일강수량': item['sumRn'],
            '평균풍속': item['avgWs'],
            '일적설량': item['ddMefs']
        }

        results_list.append(results)

    return results_list

# 테스트 코드
if __name__ == "__main__":
    results = weather_data()
    for result in results:
        print(result)
```
