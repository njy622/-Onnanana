```python
import pandas as pd
```

```python
# 데이터 불러오기
df = pd.read_csv('data/측정좌표_전처리.csv')
df.head()
```

```python
tmp_gu_dict = {
    '수원': ['장안구', '권선구', '팔달구', '영통구'], 
    '성남': ['수정구', '중원구', '분당구'], 
    '안양': ['만안구', '동안구'], 
    '안산': ['상록구', '단원구'], 
    '고양': ['덕양구', '일산동구', '일산서구'], 
    '용인': ['처인구', '기흥구', '수지구'], 
    '청주': ['상당구', '서원구', '흥덕구', '청원구'], 
    '천안': ['동남구', '서북구'], 
    '전주': ['완산구', '덕진구'], 
    '포항': ['남구', '북구'], 
    '창원': ['의창구', '성산구', '진해구', '마산합포구', '마산회원구']
}
```

```python
metro_list = ['서울특별시','부산광역시','대구광역시','인천광역시','대전광역시','광주광역시','울산광역시', '세종특별자치시', '제주특별자치도']
```

```python
si = {
    '수원시','성남시','안양시','안산시','고양시','용인시','청주시','천안시','전주시','포항시','창원시'
}
```

```python
si_name = [None] * len(df)

for i in df.index:
    if df.광역시도[i] in metro_list:
        if len(df.시군구[i]) == 2:
            si_name[i] = df.광역시도[i][:2] + ' ' + df.시군구[i]      # 서울 중구, 대전 서구
        elif df.광역시도[i] == '세종특별자치시': # 세종특별자치시
            si_name[i] = df.광역시도[i][:2]      
        elif df.광역시도[i] == '제주특별자치도': # 제주특별자치도
            si_name[i] = df.시군구[i].replace('시', '')
        elif df.시군구[i] == '군위군':
            si_name[i] = df.시군구[i][:2]
        else:
            si_name[i] = df.광역시도[i][:2] + ' ' + df.시군구[i][:-1] # 서울 용산, 서울 
    else:
        if df.시군구[i][:-1] == '고성':
            if df.광역시도[i] == '강원특별자치도':
                si_name[i] = '고성(강원)'
            else:
                si_name[i] = '고성(경남)' 
        else:
            si_name[i] = df.시군구[i][:-1]     # 광명, 의정부
            
        for key, value in tmp_gu_dict.items():
            if df.구읍[i] in value:
                if len(df.구읍[i]) == 2:
                    si_name[i] = key + ' ' + df.구읍[i]     # 포항 북구, 포함 남구
                elif len(df.구읍[i]) == 5:
                    si_name[i] = key + ' ' + df.구읍[i].replace('마산', '').replace('구', '')      # 창원 합포, 창원 회원
                else:
                    si_name[i] = key + ' ' + df.구읍[i].replace('구', '')    # 수원 장안, 고양 일산동
```

```python
print(si_name)
```

```python
df.index = si_name
df.index.name = 'ID'
df['중복확인'] = si_name
df.head()
```

```python
# csv파일로 저장
df.to_csv('data/측정좌표_ID.csv', encoding='utf-8', index=False)
```
