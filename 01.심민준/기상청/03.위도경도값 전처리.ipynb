```python
import pandas as pd
```

```python
# 데이터 불러오기
df = pd.read_csv('data/전국 위도,경도 값.csv')
df.head()
```

```python
# 필요없는 컬럼 삭제
df.drop(columns=['구분','행정구역코드','경도(시)','경도(분)','경도(초)','위도(시)','위도(분)','위도(초)','경도(초/100)','위도(초/100)','위치업데이트'], inplace=True)
# '3단계' 열의 값이 NaN이고 '2단계' 열의 값이 NaN이 아닌 행 남기기
df = df[df['3단계'].isna() & df['2단계'].notna()]
df.head()
```

```python
# 3단계 컬럼 삭제
df = df.drop(columns='3단계')
df
```

```python
df.rename(columns={'1단계': '광역시도'}, inplace=True)
df.rename(columns={'2단계': '시군구'}, inplace=True)
df
```

```python
# csv파일로 저장
df.to_csv('data/위도경도_전처리.csv', index=False, encoding='utf-8')
```
