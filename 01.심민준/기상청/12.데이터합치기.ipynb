```python
import pandas as pd
```

```python
pop = pd.read_csv('data/측정좌표_최종.csv', encoding='utf-8')
pop.head()
```

```python
df = pd.read_csv('data/카토그램.csv', encoding='utf-8')
df.head()
```

```python
# '중복확인' 열과 'ID' 열이 일치하는 경우, '측정소명' 열을 일치하는 행에 추가
df['측정소명'] = df['ID'].map(pop.set_index('중복확인')['측정소명'])
df
```

```python
# csv파일로 저장
df.to_csv('data/카토그램_최종.csv', encoding='utf-8', index=False)
```
