```python
# 완성코드

def get_air_quality(sidoName):
    
    # 웹 요청시 같이 전달될 데이터 = 요청 메시지
    params = {
            'serviceKey' : service_key,
            'returnType' : 'JSON',
            'numOfRows' : 30,
            'pageNo' : 1,
            'sidoName' : sidoName, # 시도 명
            'ver' : "1.0" # 오퍼레이션 버전
        }

    res = requests.get(url=base_url , params=params)
    # print(res.status_code, type(res.text), res.url)
    # print()
    # print(res.text)

    # 응답 데이터 정리
    from pprint import pprint  # 구조있는 데이터를 더 편하게 보여줌
    data = res.json() # json.loads(res.text)와 같은 기능
    data = data['response']['body']['items']
    
    # 최종 데이터 생성
    results = {}
    for d in data:
        results['아황산가스 농도'] = d['so2Value'] + "ppm"
        results['일산화탄소 농도'] = d['coValue'] + "ppm"
        results['오존 농도'] = d['o3Value'] + "ppm"
        results['이산화질소 농도'] = d['no2Value'] + "ppm"
        results['미세먼지(PM10) 농도'] = d['pm10Value'] + "ug/m³"
        results['초미세먼지(PM2.5) 농도'] = d['pm25Value'] + "ug/m³"
    # pprint(results)
    return results
    
# 테스트 코드
if __name__ == "__main__":
    print("서울 대기질 농도" , get_air_quality('서울'))
```
