```python
import requests, json
import pandas as pd
from urllib.parse import quote
```

```python
df = pd.read_csv("data/측정소좌표.csv", index_col="Unnamed: 0")
df
```

```python
df.주소
```

```python
df['광역시도'] = df['주소'].str.split().str[0]
df['시군구'] = df['주소'].str.split().str[1]
df['구읍'] = df['주소'].str.split().str[2]
```

```python
df.loc[df.광역시도 == '서울', '광역시도'] = '서울특별시'
df.loc[df.광역시도 == '부산', '광역시도'] = '부산광역시'
df.loc[df.광역시도 == '대구', '광역시도'] = '대구광역시'
df.loc[df.광역시도 == '인천', '광역시도'] = '인천광역시'
df.loc[df.광역시도 == '광주', '광역시도'] = '광주광역시'
df.loc[df.광역시도 == '대전', '광역시도'] = '대전광역시'
df.loc[df.광역시도 == '울산', '광역시도'] = '울산광역시'
df.loc[df.광역시도 == '세종', '광역시도'] = '세종특별자치시'
df.loc[df.광역시도 == '강원', '광역시도'] = '강원특별자치도'
df.loc[df.광역시도 == '제주', '광역시도'] = '제주특별자치도'
df.loc[df.광역시도 == '경기', '광역시도'] = '경기도'
df.loc[df.광역시도 == '충북', '광역시도'] = '충청북도'
df.loc[df.광역시도 == '충남', '광역시도'] = '충청남도'
df.loc[df.광역시도 == '전북', '광역시도'] = '전라북도'
df.loc[df.광역시도 == '전남', '광역시도'] = '전라남도'
df.loc[df.광역시도 == '경북', '광역시도'] = '경상북도'
df.loc[df.광역시도 == '경남', '광역시도'] = '경상남도'
```
