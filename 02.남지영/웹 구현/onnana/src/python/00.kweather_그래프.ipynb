```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
```

```python
df = pd.read_csv('data/kweather.csv')
```

```python
# 한글 폰트 경로 설정
font_path = "C:/Windows/Fonts/malgun.ttf"  # 적절한 한글 글꼴 경로로 변경

# 폰트 설정
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
# 테두리 색 및 두께 설정
edgecolor = 'gray'
linewidth = 1  # 두께 조절

# 도넛 그래프 그리기
for i, row in df.iterrows():
    index_value = row['그래프 값']
    sizes = [index_value, row['그래프 최댓값'] - index_value]

    # 도넛 그래프 설정
    fig, ax = plt.subplots()

    # 6시에서 시작하여 9시 지점으로 도넛 그래프 그리기
    wedges, texts, autotexts = ax.pie(sizes, labels=None, autopct='',
                                      startangle=-90, textprops=dict(color="w"),
                                      wedgeprops=dict(width=0.1, edgecolor=edgecolor, linewidth=linewidth, facecolor='none'))  # facecolor 추가

    # 지수에 해당하는 그래프 색상 지정 (하늘색)
    wedges[0].set_facecolor('#87CEEB')  # 스카이 블루 (하늘색)

    # 원그래프를 원형으로 유지
    ax.axis('equal')

    # 그래프 표시
    plt.text(0, -1.3, f"{row['생활지수']}\n{row['그래프 값']}", ha='center', va='center', fontsize=12, fontweight='bold', color='black')
    plt.show()
```
