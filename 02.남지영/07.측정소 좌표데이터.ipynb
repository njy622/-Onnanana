```python
# Label Encoding, ['ham', 'spam']  --> [0, 1] 로 변경
df.v1 = df.v1.replace(['ham', 'spam'], [0, 1])      # Lable Encoding을 이용해서 변경해도 상관없음
df.head(3)
```

```python
# Ham/Spam 분포
df.v1.value_counts()
```

- 텍스트 전처리

```python
# 구둣점, 숫자 제거
df.v2 = df.v2.str.replace('[^A-Za-z]', ' ', regex=True)
df.v2[0]
```

- 데이터셋 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.v2.values, df.v1.values, stratify=df.v1.values, test_size=0.2, random_state=2023
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

- Text Encoding

```python
from sklearn.feature_extraction.text import CountVectorizer
cvect = CountVectorizer(stop_words='english')
```

```python
# 이렇게 하면 절대로 안됨
# train/test dataset에서 사용되는 단어가 다름(같은 숫자로 되어야함!) 
X_train_cv = cvect.fit_transform(X_train)
X_test_cv = cvect.fit_transform(X_test)
X_train_cv.shape, X_test_cv.shape

# 결과값이 ((4135, 6494), (1034, 2866)) 서로 다름
```

```python
# 변환된 단어 세트가 train과 test가 동일해야함
cvect.fit(X_train)
X_train_cv = cvect.transform(X_train)
X_test_cv = cvect.transform(X_test)

X_train_cv.shape, X_test_cv.shape
```

- 학습 및 평가

```python
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression(random_state=2023)
```

```python
lrc.fit(X_train_cv, y_train)
```

```python
lrc.score(X_test_cv, y_test)
```
