## 탄소계산기 함수

탄소발자국 계산 공식
마지막 업데이트 날짜: 2021-03-17

탄소발자국 계산 공식은 측정하는 항목에 따라 다릅니다.

에너지에 대한 탄소발자국 계산
에너지 탄소배출량 계산에는 선택한 기본 측정 단위가 사용됩니다.

범위 1(직접) 에너지 배출에는 다음과 같은 계산이 사용됩니다.

CO2 = (총량 * (CO2 배출 계수 * 난방 값) * 밀도)

CH4 = ((총량 * (CH4 배출 계수 * 난방 값) * 밀도) * GWP CH4 환산)

N2O = ((총량 * (N2O 배출 계수 * 난방 값) * 밀도) * GWP N2O 환산)

범위 2(간접) 에너지 배출에는 다음과 같은 계산이 사용됩니다.

CO2 = ((점유 기준의 총 사용량 * 배출 계수))

CH4 = ((점유 기준의 총 사용량 * CH4 배출 계수) * GWP CH4 환산)

N2O = ((점유 기준의 총 사용량 * N2O 배출 계수) * GWP N2O 환산)

이동에 대한 탄소발자국 계산
회사가 차량을 소유한 경우 배출량은 이동한 거리로 계산됩니다. 회사가 차량을 소유하지 않은 경우 배출량은 사용으로 계산됩니다.

회사 소유 차량에 대한 범위 1(직접) 이동 배출에는 다음과 같은 계산이 사용됩니다.

CO2 = ((이동 거리 + 연료 사용) * 배출 계수)

CH4 = (((이동 거리 + 연료 사용) * CH4 배출 계수) * GWP CH4 환산)

N2O = (((이동 거리 + 연료 사용) * N2O 배출 계수) * GWP N2O 환산)

회사에서 소유하지 않은 차량에 대한 범위 1(직접) 이동 배출에는 다음과 같은 계산이 사용됩니다.

CO2 = ((이동 거리 + 연료 사용) * (배출 계수 * 난방 값) * 밀도)

CH4 = (((이동 거리 + 연료 사용) * (CH4 배출 계수 * 난방 값) * 밀도) * GWP CH4 Conversion)

N2O = (((이동 거리 + 연료 사용) * (N2O 배출 계수 * 난방 값) * 밀도)*GWP N2O 환산)

회사 소유 차량에 대한 범위 3(간접) 이동 배출에는 다음과 같은 계산이 사용됩니다.

이동 탄소발자국 = (배출 계수 * 이동 거리)

회사에서 소유하지 않은 차량에 대한 범위 3(간접) 이동 배출에는 다음과 같은 계산이 사용됩니다.

이동 탄소발자국 = ((이동 거리 + 연료 사용) * 배출 계수 * 난방 값 * 밀도)

기타 배출에 대한 탄소발자국 계산
기타 배출에는 다음과 같은 계산이 사용됩니다.

기타 배출 = 총 수량 * 0.001102 * GWP 배출 계수
