```python
import requests

def get_current_location():
    # 카카오 REST API 호출을 위한 URL
    url = "https://dapi.kakao.com/v2/local/geo/coord2regioncode.json"

    # 현재 위치 정보를 받아오기 위한 요청
    headers = {
        "Authorization": "8b02b6faecb38823b8d8180f058462a1"  # 본인의 REST API 키 입력
    }

    try:
        # 현재 위치를 기준으로 지역코드 및 좌표 정보를 받아옴
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            result = response.json()
            x = result['documents'][0]['x']  # 경도
            y = result['documents'][0]['y']  # 위도
            return x, y
        else:
            print(f"요청에 실패했습니다. 상태 코드: {response.status_code}")
            return None
    except Exception as e:
        print(f"오류 발생: {e}")
        return None

# 현재 위치의 좌표 받아오기
current_location = get_current_location()
if current_location:
    print(f"현재 위치 좌표: {current_location[0]}, {current_location[1]}")
```
